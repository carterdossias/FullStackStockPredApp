```python
import numpy as np
import pandas as pd
import mysql.connector
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler

# Database connection parameters
db_config = {
    'host': ipCred,
    'user': usernameCred,
    'password': passwordCred,
    'database': databaseCred
}

# Get ticker from user
ticker = input("Enter the ticker symbol (e.g., AAPL, ORCL): ").strip().upper()

# Establish connection
conn = mysql.connector.connect(**db_config)
query = f"""
    SELECT date, open, high, low, close, volume, total_sentiment
    FROM {ticker}_view
    WHERE date BETWEEN '2024-03-14' AND '2025-3-13'
    ORDER BY date ASC;
"""
df = pd.read_sql(query, conn)
conn.close()

# Convert date column
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Check for NaNs and fill missing values
if df.isnull().values.any():
    print("Warning: NaN values detected in dataset! Filling missing values...")
    df.fillna(method='ffill', inplace=True)  # Forward-fill missing data

# Selecting features
features = ['open', 'high', 'low', 'close', 'volume', 'total_sentiment']
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df[features])

# Function to create sequences for LSTM
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length, 3])  # Predicting 'close' price
    return np.array(X), np.array(y)

# Reduce sequence length for better learning
SEQ_LENGTH = 30  # Reduced from 60
X, y = create_sequences(df_scaled, SEQ_LENGTH)

# Train-test split (90% train, 10% test)
split = int(0.9 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Define LSTM Model with gradient clipping
model = Sequential([
    LSTM(100, return_sequences=True, input_shape=(SEQ_LENGTH, len(features))),
    Dropout(0.2),
    LSTM(100, return_sequences=False),
    Dropout(0.2),
    Dense(50, activation='relu'),
    Dense(1)
])

# Use Adam optimizer with gradient clipping
optimizer = Adam(learning_rate=0.001, clipnorm=1.0)
model.compile(optimizer=optimizer, loss='mae')

# Train the model with increased epochs & reduced batch size
model.fit(X_train, y_train, epochs=200, batch_size=8, validation_data=(X_test, y_test))

# Predictions
y_pred = model.predict(X_test)

# Function to denormalize values
def denormalize(value, scaler, column_index):
    base = np.zeros((value.shape[0], len(features)))  # Create placeholder
    base[:, column_index] = value.flatten()  # Assign predicted values
    return scaler.inverse_transform(base)[:, column_index]

# Denormalize predictions
y_test_actual = denormalize(y_test, scaler, 3)
y_pred_actual = denormalize(y_pred, scaler, 3)

# Debugging check for NaNs
if np.isnan(y_pred_actual).any():
    print("Error: Predicted values contain NaNs!")

# Plot results
plt.figure(figsize=(12,6))
plt.plot(df.index[split + SEQ_LENGTH:], y_test_actual, label="Actual Closing Price", color='blue')
plt.plot(df.index[split + SEQ_LENGTH:], y_pred_actual, label="Predicted Closing Price", color='red')
plt.title(f"{ticker} Stock Price Prediction")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

# Print first 10 predictions for verification
print("First 10 Predicted Closing Prices:", y_pred_actual[:10])
```
